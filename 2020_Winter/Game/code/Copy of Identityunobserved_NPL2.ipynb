```
import numpy as np
import random as rn
import itertools as its
from sklearn.preprocessing import normalize
from tqdm import tqdm

class DataGeneration(object):
    def __init__(self, theta ,N,T,x0_dim):
      self.reward = theta #s a1 a2
      self.x0_dim = x0_dim

      self.dygame = np.zeros([2,2,2,2,2]) #st,at1, at2,st+1, zt+1
      ind = 0
      for st, at1, at2 in its.product(range(2),range(2),range(2)):
        ind = ind + 1
        rn.seed(ind)
        prob = rn.choices(np.arange(11), weights=None, cum_weights=None, k=4)
        self.dygame[st,at1,at2,:,:] = np.reshape(prob/np.sum(prob),[2,2])
      
      self.N = N
      self.T = T
      
      self.gamma = 0.5772
      self.beta = 0.9

      self.blief = 31
      self.iniccp = np.ones([self.blief,2])/2#np.vstack([np.ones(self.blief),np.zeros(self.blief)]).T

    def rw(self,s,a1,a2,theta2): # sc in {0,1}, a1,a2 in {0,1}
      theta2 = np.reshape(theta2,[2,2,2])
      return theta2[s,a1,a2]

    def siglam(self,z,x,a1,a2,theta1):
      theta1 = np.reshape(theta1,[2,2,2,2,2])
      lambd = theta1[:,a1,a2,:,z].T.dot([x,1-x])
      sigma = np.sum(lambd)
      return lambd[0]/sigma, sigma

    def PartCCP(self,ccp,ccpc,theta2,theta1, maxiter,thread1):
      Q = np.zeros([self.blief,2])
      #print(maxiter)    
      for j in range(maxiter):
        Q_old = Q.copy()
        V = self.gamma + np.log(np.sum(np.exp(Q),axis=1))
        
        for indx, inda in its.product(self.x,range(2)):
          
          l = np.zeros([2,2])#z,a1
          s = np.zeros([2,2])
          V_tep = np.zeros([2,2])
          for indl, inds in its.product(range(2),range(2)): #z , a2
            
            l[indl,inds],s[indl,inds] = (ccp==2)*self.siglam(indl,indx,inds,inda,theta1)+\
                                        (ccp==1)*self.siglam(indl,indx,inda,inds,theta1) #z , a1
            fl = int(np.floor(l[indl,inds] * (self.blief-1)))
            if fl == self.blief-1:
              V_tep[indl,inds] = V[fl]
            else: 
              V_tep[indl,inds] = V[fl] * (l[indl,inds]*(self.blief-1) - fl) + V[fl+1]* (1 + fl -l[indl,inds]*(self.blief-1)) 
          
          V_tep = V_tep * s
          
          ind = int(indx *(self.blief-1))
                          
          Q[ind,inda] = (ccp==2)*(ccpc[ind,0] *( indx* self.rw(0,0,inda,theta2) + (1-indx) * self.rw(1,0,inda,theta2) + self.beta * np.sum(V_tep[:,0])) +\
                          ccpc[ind,1] * (indx* self.rw(0,1,inda,theta2) + (1-indx) * self.rw(1,1,inda,theta2) + self.beta * np.sum(V_tep[:,1]))) +\
                        (ccp==1)*(ccpc[ind,0] *( indx* self.rw(0,inda,0,theta2) + (1-indx) * self.rw(1,inda,0,theta2) + self.beta * np.sum(V_tep[:,0])) +\
                           ccpc[ind,1] * (indx* self.rw(0,inda,1,theta2) + (1-indx) * self.rw(1,inda,1,theta2) +self.beta * np.sum(V_tep[:,1])))

        if np.max(np.abs(Q-Q_old))<=thread1:
          break
      ccpcc = normalize(np.exp(Q), norm='l1',axis=1)#np.exp(Q1)/(np.sum(np.exp(Q1),axis=1)) 
      return ccpcc

    def PerfectMarkov(self,ccp,iniCCP1,theta2,theta1, maxiter = 100,thread1 = 0.01,thread2 = 1e-4):      
      self.x = np.linspace(0,1,self.blief)     
      ccp1 = iniCCP1
      ccp2 = iniCCP1
      #for i in tqdm(range(maxiter)):
      for i in range(maxiter):

        ccp1_old = ccp1.copy()
        ccp2_old = ccp2.copy()
        if ccp==1:
          ccp2 = self.PartCCP(ccp+1,ccp1,theta2,theta1,maxiter,thread1)
          ccp1 = self.PartCCP(ccp,ccp2,theta2,theta1,maxiter,thread1)
        else:
          ccp1 = self.PartCCP(ccp-1,ccp2,theta2,theta1,maxiter,thread1)
          ccp2 = self.PartCCP(ccp,ccp1,theta2,theta1,maxiter,thread1)
       
        # Q2 = np.zeros([self.blief,2])
        
        # for j in range(maxiter):
        #   Q2_old = Q2.copy()
        #   V2 = self.gamma + np.log(np.sum(np.exp(Q2),axis=1))
          
        #   for indx, inda in its.product(x,range(2)):
            
        #     l2 = np.zeros([2,2])#z,a1
        #     s2 = np.zeros([2,2])
        #     V2_tep = np.zeros([2,2])
        #     for indl, inds in its.product(range(2),range(2)): #z , a2
              
        #       l2[indl,inds],s2[indl,inds] = self.siglam(indl,indx,inds,inda,theta1) #z , a1
        #       fl = int(np.floor(l2[indl,inds] * (self.blief-1)))
        #       if fl == self.blief-1:
        #         V2_tep[indl,inds] = V2[fl]
        #       else: 
        #         V2_tep[indl,inds] = V2[fl] * (l2[indl,inds]*(self.blief-1) - fl) + V2[fl+1]* (1 + fl -l2[indl,inds]*(self.blief-1)) 
            
        #     V2_tep = V2_tep * s2
            
        #     ind = int(indx *(self.blief-1))
                            
        #     Q2[ind,inda] = ccp1[ind,0] *( indx* self.rw(0,0,inda,theta2) + (1-indx) * self.rw(1,0,inda,theta2) + self.beta * np.sum(V2_tep[:,0])) +\
        #                     ccp1[ind,1] * (indx* self.rw(0,1,inda,theta2) + (1-indx) * self.rw(1,1,inda,theta2) + self.beta * np.sum(V2_tep[:,1]))

        #   if np.max(np.abs(Q2-Q2_old))<=thread1:
        #     break
        # ccp2 = normalize(np.exp(Q2), norm='l1',axis=1)#np.exp(Q1)/(np.sum(np.exp(Q1),axis=1))
        
        
        # Q1 = np.zeros([self.blief,2])        
        # for j in range(maxiter):
        #   Q1_old = Q1.copy()
        #   #Q2_old = Q2.copy()
          
        #   V1 = self.gamma + np.log(np.sum(np.exp(Q1),axis=1))
        #   for indx, inda in its.product(x,range(2)):
            
        #     l1 = np.zeros([2,2])#z,a2
        #     s1 = np.zeros([2,2])
        #     V1_tep = np.zeros([2,2])

        #     for indl, inds in its.product(range(2),range(2)): #z , a2
        #       ##print(siglam(indl,indx,inda,inds))
        #       l1[indl,inds],s1[indl,inds] = self.siglam(indl,indx,inda,inds,theta1)
        #       fl = int(np.floor(l1[indl,inds] * (self.blief-1)))
        #       if fl == self.blief-1:
        #        V1_tep[indl,inds] = V1[fl]
        #       else: 
        #        V1_tep[indl,inds] = V1[fl] * (l1[indl,inds]*(self.blief-1) - fl) + V1[fl+1]* (1 + fl -l1[indl,inds]*(self.blief-1)) 
            
        #     #print(indl)
        #     V1_tep = V1_tep * s1

        #     ind = int(indx *(self.blief-1))
        #     Q1[ind,inda] = ccp2[ind,0] *( indx* self.rw(0,inda,0,theta2) + (1-indx) * self.rw(1,inda,0,theta2) + \
        #                      self.beta * np.sum(V1_tep[:,0])) +\
        #                    ccp2[ind,1] * (indx* self.rw(0,inda,1,theta2) + (1-indx) * self.rw(1,inda,1,theta2) +\
        #                      self.beta * np.sum(V1_tep[:,1]))
                            
        #   #print(j,haha1)
        #   if np.max(np.abs(Q1-Q1_old))<=thread1:
        #     #print(i,j)
        #     #print(j,np.max(Q1),np.max(Q2),np.max(np.abs(Q1-Q1_old)),np.max(np.abs(Q2-Q2_old)))
        #     break
        #   #print(j,np.max(Q1),np.max(Q2),np.max(np.abs(Q1-Q1_old)),np.max(np.abs(Q2-Q2_old)))
        # ccp1 = normalize(np.exp(Q1), norm='l1',axis=1)#np.exp(Q1)/(np.sum(np.exp(Q1),axis=1))
        # #ccp2 = normalize(np.exp(Q2), norm='l1',axis=1)#np.exp(Q1)/(np.sum(np.exp(Q1),axis=1))


        if np.max(np.abs(ccp1-ccp1_old))<=thread2 and np.max(np.abs(ccp2-ccp2_old))<=thread2:
          #print(i,ccp1,ccp2)
          #print(i,np.max(ccp1),np.max(ccp2),np.max(np.abs(ccp1-ccp1_old)),np.max(np.abs(ccp2-ccp2_old)))
          break
        #print(i,ccp1,ccp2)  
        #print(i,np.max(ccp1),np.max(ccp2),np.max(np.abs(ccp1-ccp1_old)),np.max(np.abs(ccp2-ccp2_old)))
      if i == maxiter -1:
        print('fail to converge')
      return ccp1, ccp2, i


    def simulation(self):
      ccp1, ccp2,_ = self.PerfectMarkov(1,self.iniccp,self.reward,self.dygame)

      self.st = np.zeros([self.N,self.T+1],dtype = int)
      self.zt = np.zeros([self.N,self.T],dtype = int)
      self.a1t = np.zeros([self.N,self.T],dtype = int)
      self.a2t = np.zeros([self.N,self.T],dtype = int)
      self.belief = 1/2 * np.ones([self.N,self.T+1])
      self.belief[:,0] = np.kron(np.ones(int(self.N/self.x0_dim)),np.linspace(0,1,num = self.x0_dim))

      rn.seed(0)
      self.st[:,0] = rn.choices([0,1],[1/2,1/2], k = self.N)

      for ss in range(self.N):
        for t in range(self.T):
          fl = int(np.floor(self.belief[ss,t] * (self.blief-1)))
          if fl == self.blief-1:
            prob1 = ccp1[fl,:]
            prob2 = ccp2[fl,:]
          else: 
            prob1 = ccp1[fl,:] * (self.belief[ss,t] *(self.blief-1) - fl) + ccp1[fl+1,:]* (1 + fl -self.belief[ss,t]*(self.blief-1))
            prob2 = ccp2[fl,:] * (self.belief[ss,t] *(self.blief-1) - fl) + ccp2[fl+1,:]* (1 + fl -self.belief[ss,t]*(self.blief-1))
          
          self.a1t[ss,t] = rn.choices([0,1],prob1)[0]
          self.a2t[ss,t] = rn.choices([0,1],prob2)[0]

          prob3 = self.dygame[self.st[ss,t], self.a1t[ss,t],self.a2t[ss,t],:,:]#st,at1, at2,st+1, zt+1

          self.st[ss,t+1], self.zt[ss,t] = rn.choices([[0,0],[0,1],[1,0],[1,1]], np.reshape(prob3,[4,1]))[0]
          self.belief[ss,t+1],_ = self.siglam(self.zt[ss,t],self.belief[ss,t],self.a1t[ss,t],self.a2t[ss,t],self.dygame)
      return {'state':self.st, 'observation':self.zt, 'action1':self.a1t, 'action2': self.a2t,'belief':self.belief}
```
